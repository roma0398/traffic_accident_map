```python
import re
import pandas as pd
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
```

```python
data = pd.read_csv('data/post_data.csv')
data = data.dropna()
data = data.reset_index()
```

```python
data.drop(['index'], axis=1, inplace=True)
```

```python
df = data[:1000]
df.head()
```

```python
result = re.search('(.*)(–ù–∞ | –Ω–∞ |—É–ª–∏—Ü|–ø–µ—Ä–µ—Å–µ—á–µ–Ω|–ø–µ—Ä–µ–∫—Ä–µ—Å—Ç|–ø—Ä–æ—Å–ø–µ–∫—Ç)(.*)', df['text'][1])
print(result.group(3))
```

```python
dop=[]
ind=[]
a=[]
k = 0
pat = re.compile('(–ù–∞ | –Ω–∞ |—É–ª–∏—Ü|–ø–µ—Ä–µ—Å–µ—á–µ–Ω|–ø–µ—Ä–µ–∫—Ä–µ—Å—Ç|–ø—Ä–æ—Å–ø–µ–∫—Ç)(.*)')
for i in list(df['text']):
    if (pat.search(i) is not None):
        a.append(pat.search(i).group(2))
        dop.append(pat.search(i).group(1))
        ind.append(df['id'][k])
    k+=1
```

```python
titleList = []
for str in a:
    tit = []
    words = word_tokenize(str)
    for word in words:
        if word.istitle() and len(word)>3:
            tit.append(word)
            
    titleList.append(tit)
    
```

```python
print(titleList[:15])
```

```python
streets = set(line.strip() for line in open('data/street_list_spb.txt'))
```

```python
stemmer = SnowballStemmer("russian") 
stemmer.stem("–°–æ—Ñ–∏–π—Å–∫—É—é")
```

```python
finded=[]
for titl in titleList:
    q=[]
    for s in titl:
        q.append(s)
        stemmed = stemmer.stem(s)
        for val in streets:
            if stemmed in val.lower():
                q.append(val)
    finded.append(q)         
        
```

```python
finded=[]
for titl in titleList:
    dict = {}
    for s in titl:
        w=[]
        stemmed = stemmer.stem(s)
        for val in streets:
            if stemmed in val.lower():
                w.append(val)
        if w:
            dict[s]=w
    finded.append(dict)         
        
```

```python
finded[:15]
```

```python
test = finded[:15]
ans =[]
for i in range(len(test)):
    d=test[i]
    flat_list = [item for sublist in d.values() for item in sublist]
    intersec= set(flat_list)
    for key in d:
        intersec.intersection_update(set(d[key]))
    if len(intersec)==1:
        test[i]={'ans':list(intersec)}
    ans.append(intersec)
        
print(ans)
```
